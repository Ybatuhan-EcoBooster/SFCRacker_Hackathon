```python
def extract_paragraph(title):
        
        # Use regex to find the title and its corresponding paragraph
        pattern = re.compile(fr'{title}\s*([\s\S]+?)(?=\n\n|$)')
        match = pattern.search(text)

        if match:
            paragraph = match.group(1).strip()
            return paragraph
        else:
            return f"Title '{title}' not found in the file."

text_list = []

for title in unmatched_data:
    result = extract_paragraph(title)
    text_list.append(result)

text_list = pd.DataFrame(text_list)
text_list['NAZWA SEKCJI (WEDŁUG ROZPORZĄDZENIA)'] = unmatched_data
text_list.rename(columns={0:'Text'},inplace=True)
text_list

```
